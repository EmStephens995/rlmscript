# Linear Modeling in R

```r
df = read.csv("regrex1.csv")
```

```r
head(df)
```

```r
plot(df$x,df$y,
main = "Scatter Plot",
xlab = "X Axis",
ylab = "Y Axis")
```

```r
model = lm(y~x, data = df)
```

```r
summary_stats = summary (model)
```

```r
summary_stats$r.squared
```

```r
coef(model)
```

```r
coef(model)[1]
```

```r
coef(model)[2]
```

```r
y_pred = predict(model, df)
```

```r
plot(df$x,y_pred)
```

```r
plot(df$x,df$y, 
     main = "Linear Regression",
     xlab = "X Axis",
     ylab = "Y Axis")
lines(df$x,y_pred)
```
